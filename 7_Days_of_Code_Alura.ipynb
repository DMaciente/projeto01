### Importando Bibliotecas

```python
import pandas as pd
import os
```

### Gerando DATAFRAME com emprestimos
df_emprestimo

```python
url_base = 'https://raw.githubusercontent.com/FranciscoFoz/7_Days_of_Code_Alura-Python-Pandas/main/Dia_1-Importando_dados/Datasets/dados_emprestimos/'
```

```python
arquivos = ['emprestimos-20101.csv', 'emprestimos-20102.csv', 'emprestimos-20111.csv',
            'emprestimos-20112.csv', 'emprestimos-20121.csv', 'emprestimos-20122.csv',
            'emprestimos-20131.csv', 'emprestimos-20132.csv', 'emprestimos-20141.csv',
            'emprestimos-20142.csv', 'emprestimos-20151.csv', 'emprestimos-20152.csv',
            'emprestimos-20161.csv', 'emprestimos-20162.csv', 'emprestimos-20171.csv',
            'emprestimos-20172.csv', 'emprestimos-20181.csv', 'emprestimos-20182.csv',
            'emprestimos-20191.csv', 'emprestimos-20192.csv', 'emprestimos-20201.csv']
```

```python
df_emprestimo = pd.DataFrame()

for arquivo in arquivos:
  url = url_base + arquivo
  df = pd.read_csv(url)
  df_emprestimo = pd.concat([df_emprestimo, df])

df_emprestimo = df_emprestimo.drop_duplicates()
```

### Gerando DATAFRAME com exemplares
df_exemplar

```python
 df_exemplar = pd.read_parquet('https://raw.githubusercontent.com/FranciscoFoz/7_Days_of_Code_Alura-Python-Pandas/main/Dia_1-Importando_dados/Datasets/dados_exemplares.parquet')
 df_exemplar.head()
```

### Juntando os DATAFRAMES

```python
df_final = pd.merge(df_emprestimo, df_exemplar, on='codigo_barras')
```

```python
df_final
```

### ETL

```python
# excluíndo coluna 'registro de sistema'
df_final = df_final.drop(columns=['registro_sistema'])
```

```python
# transformando coluna 'matricula_ou_siape' em string
df_final['matricula_ou_siape'] = df_final['matricula_ou_siape'].astype(str)
```

```python
# incluíndo coluna 'classificacao' do sistema CDU - Classificação Decimal Universal com base na coluna localização
# 000 a 099: Generalidades. Ciência e conhecimento.
# 100 a 199: Filosofia e psicologia.
# 200 a 299: Religião.
# 300 a 399: Ciências sociais.
# 400 a 499: Classe vaga. Provisoriamente não ocupada.
# 500 a 599: Matemática e ciências naturais.
# 600 a 699: Ciências aplicadas.
# 700 a 799: Belas artes.
# 800 a 899: Linguagem. Língua. Linguística.
# 900 a 999: Geografia. Biografia. História.

df_final.loc[(df_final['localizacao'] >= 0) & (df_final['localizacao'] <= 99), 'classificacao'] = 'Generalidades. Ciência e conhecimento.'
df_final.loc[(df_final['localizacao'] >= 100) & (df_final['localizacao'] <= 199), 'classificacao'] = 'Filosofia e psicologia.'
df_final.loc[(df_final['localizacao'] >= 200) & (df_final['localizacao'] <= 299), 'classificacao'] = 'Religião.'
df_final.loc[(df_final['localizacao'] >= 300) & (df_final['localizacao'] <= 399), 'classificacao'] = 'Ciências sociais.'
df_final.loc[(df_final['localizacao'] >= 400) & (df_final['localizacao'] <= 499), 'classificacao'] = 'Classe vaga. Provisoriamente não ocupada.'
df_final.loc[(df_final['localizacao'] >= 500) & (df_final['localizacao'] <= 599), 'classificacao'] = 'Matemática e ciências naturais.'
df_final.loc[(df_final['localizacao'] >= 600) & (df_final['localizacao'] <= 699), 'classificacao'] = 'Ciências aplicadas.'
df_final.loc[(df_final['localizacao'] >= 700) & (df_final['localizacao'] <= 799), 'classificacao'] = 'Belas artes.'
df_final.loc[(df_final['localizacao'] >= 800) & (df_final['localizacao'] <= 899), 'classificacao'] = 'Linguagem. Língua. Linguística.'
df_final.loc[(df_final['localizacao'] >= 900) & (df_final['localizacao'] <= 999), 'classificacao'] = 'Geografia. Biografia. História.'
```
